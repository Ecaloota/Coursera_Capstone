```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.auhouseprices.com/act/"
req = requests.get(url)
soup = BeautifulSoup(req.text, 'lxml')

outlist = []
for tag in soup.find_all("li", {"class": "list-group-item"}):
    outlist.append(tag.text)

    
x = outlist[0:48]+outlist[72:]
for elem in x:
    print(elem)
```
